This notebook provides analytical insights on the top 100 cryptocurrencies extracted and cleaned by our ETL pipeline. It demonstrates how to fetch the processed datasets and generate static charts for business reports.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Set custom theme for visual charts
sns.set_theme(style="whitegrid")
plt.rcParams.update({
    'font.size': 11,
    'axes.labelsize': 12,
    'axes.titlesize': 14,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10
})
```

## 1. Load the Processed Data

We load the latest snapshot outputted by the transformation stage.

```python
csv_path = Path('../data/processed/latest.csv')
if not csv_path.exists():
    # Fallback to local subdirectory if running from notebooks folder
    csv_path = Path('data/processed/latest.csv')
    
df = pd.read_csv(csv_path)
print(f"Dataset loaded successfully. Shape: {df.shape}")
df.head()
```

## 2. Descriptive Market Capitalization Categories

We segment our coins into Large-cap, Mid-cap, and Small-cap brackets.

```python
category_stats = df.groupby('market_cap_category').agg(
    coin_count=('coin_id', 'count'),
    avg_price=('current_price_usd', 'mean'),
    avg_volatility=('volatility_24h', 'mean'),
    avg_liquidity_ratio=('volume_to_market_cap_ratio', 'mean')
).reset_index()

category_stats
```

## 3. Visualization 1: Top 10 Gainers in the last 24h

```python
top_gainers = df.nlargest(10, 'price_change_percentage_24h')

plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_gainers, 
    x='price_change_percentage_24h', 
    y='name', 
    palette='summer'
)
plt.title('Top 10 Performers by 24h Price Change Percentage (%)')
plt.xlabel('Change (%)')
plt.ylabel('Cryptocurrency')
plt.tight_layout()
plt.show()
```

## 4. Visualization 2: Volatility vs. Liquidity Ratio

Here we visualize whether highly volatile coins also show higher trading liquidity (volume-to-market-cap ratio).

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(
    data=df, 
    x='volume_to_market_cap_ratio', 
    y='volatility_24h', 
    hue='market_cap_category', 
    palette='Set1',
    s=90,
    alpha=0.85
)
plt.title('Cryptocurrency Analysis: Volatility vs. Liquidity Ratio')
plt.xlabel('Volume-to-Market-Cap Ratio (Liquidity)')
plt.ylabel('Volatility Index (%)')
plt.legend(title='Market Cap Category')
plt.tight_layout()
plt.show()
```
